```python
y = df_train.Makine_Sol_Zincir_Hizi
X = df_train.drop(["Makine_Sol_Zincir_Hizi"], axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, train_size=0.8, random_state=0)
```

```python
numerical_cols = [cname for cname in X_train.columns
                  if X_train[cname].dtype in ["int64", "float64"]]
```

```python
len(numerical_cols)
```

```python
categorical_cols = [cname for cname in X_train.columns
                  if X_train[cname].nunique()<10 and X_train[cname].dtype=="object"]
```

```python
len(categorical_cols)
```

```python
my_cols = numerical_cols + categorical_cols
```

```python
X_train = X_train[my_cols]
X_val = X_val[my_cols]
X_test = df_test[my_cols]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
numerical_transformer = Pipeline(steps=[
    ("impputer_num", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])
```

```python
categorical_transofmer = Pipeline(steps=[
    ("imputer_cal", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])
```

```python
from sklearn.compose import ColumnTransformer
```

```python
preprocessor = ColumnTransformer(transformers=[
    ("num", numerical_transformer, numerical_cols),
    ("cat", categorical_transofmer, categorical_cols)
])
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=0)
```

```python
my_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", rf)
])
```

```python
my_pipeline.fit(X_train, y_train)
```

```python
val_preds = my_pipeline.predict(X_val)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
print("Val MAE:", mean_absolute_error(y_val, val_preds))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = -1 *  cross_val_score(my_pipeline, X,y, cv=5, scoring="neg_mean_absolute_error")
```
